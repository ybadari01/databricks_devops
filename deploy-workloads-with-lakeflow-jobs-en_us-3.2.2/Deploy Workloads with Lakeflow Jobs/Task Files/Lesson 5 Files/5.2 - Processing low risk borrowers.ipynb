```python
%sql
CREATE OR REPLACE TABLE low_risk_borrowers_silver AS
SELECT
  id,
  has_cd_account,
  has_credit_card,
  has_personal_loan,
  is_online_customer,
  CASE
    WHEN is_online_customer = true THEN 'Digital'
    ELSE 'Branch'
  END AS preferred_channel,
  CASE
    WHEN has_cd_account = true THEN 'Saver'
    ELSE 'Non-Saver'
  END AS cd_profile
FROM loan_details_silver
WHERE NOT (has_credit_card = true AND has_personal_loan = true)
```
