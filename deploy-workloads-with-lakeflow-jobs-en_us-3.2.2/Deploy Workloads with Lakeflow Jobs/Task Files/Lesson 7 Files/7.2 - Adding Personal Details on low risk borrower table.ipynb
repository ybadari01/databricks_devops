```python
%sql

CREATE OR REPLACE TABLE low_risk_borrowers_gold AS
SELECT
  lr.id,
  pd.age,
  pd.yoe,
  pd.income_in_usd,
  pd.zip_code,
  pd.family_size,
  pd.education,
  lr.has_cd_account,
  lr.has_credit_card,
  lr.has_personal_loan,
  lr.is_online_customer,
  lr.preferred_channel,
  lr.cd_profile,

  -- Customer Type (for marketing segmentation)
  CASE
    WHEN lr.has_credit_card = true AND lr.has_personal_loan = false THEN 'Credit Only'
    WHEN lr.has_credit_card = false AND lr.has_personal_loan = true THEN 'Loan Only'
    ELSE 'Other'
  END AS customer_type,

  -- Family Size Band
  CASE
    WHEN pd.family_size <= 2 THEN 'Small'
    WHEN pd.family_size <= 4 THEN 'Medium'
    ELSE 'Large'
  END AS family_size_segment

FROM low_risk_borrowers_silver lr
JOIN borrower_details_silver pd
  ON lr.id = pd.id;
```
