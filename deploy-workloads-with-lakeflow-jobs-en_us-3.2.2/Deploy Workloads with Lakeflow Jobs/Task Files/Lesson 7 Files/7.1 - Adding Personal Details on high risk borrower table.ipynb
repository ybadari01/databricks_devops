```python
%sql
CREATE OR REPLACE TABLE high_risk_borrowers_gold AS
SELECT
  hr.id,
  pd.age,
  pd.yoe,
  pd.income_in_usd,
  pd.zip_code,
  pd.family_size,
  pd.education,
  hr.avg_cc_spending,
  hr.total_mortgage_amount,
  hr.has_cd_account,
  hr.is_online_customer,
  hr.has_securities_account,
  hr.spending_category,

  -- Income Band
  CASE
    WHEN pd.income_in_usd >= 100000 THEN 'High Income'
    WHEN pd.income_in_usd >= 50000 THEN 'Middle Income'
    ELSE 'Low Income'
  END AS income_band,

  -- Age Group
  CASE
    WHEN pd.age < 30 THEN 'Young'
    WHEN pd.age < 50 THEN 'Middle-aged'
    ELSE 'Senior'
  END AS age_group

FROM high_risk_borrowers_silver hr
JOIN borrower_details_silver pd
  ON hr.id = pd.id;
```
